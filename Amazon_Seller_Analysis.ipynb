```python
df_alter['sellerproductcount'] = df_alter['sellerproductcount'].astype(str)
```

```python
# Function to extract product count range
def extract_product_count(text):
    matches = re.findall('(\d-\d\d)', text)
    return matches[0] if matches else None

# Create a new column 'product_count' to store range
df_alter['product_count_range'] = df_alter['sellerproductcount'].apply(extract_product_count)

# Display the DataFrame
df_alter.head()
```

```python
# Function to extract product count
def extract_numerical_values(text):
    pattern = r'(\d{1,3}(?:,\d{3})*|\d+)[^\d]*results'
    matches = re.findall(pattern, text)
    return matches[0] if matches else None

# Creating a new column product_count to store number of products
df_alter['product_count'] = df_alter['sellerproductcount'].apply(extract_numerical_values)

df_alter.head()
```

### Cleaning Seller Rating Column

```python
df_alter['sellerratings'] = df_alter['sellerratings'].astype(str)
```

```python
def extract_percentage(text):
    if pd.isnull(text):
        return None
    percentage = re.findall(r'(\d+)%', text)
    return int(percentage[0]) if percentage else None

df_alter['positive_percentage_ratings'] = df_alter['sellerratings'].apply(extract_percentage)
```

```python
def extract_time(text):
    time = re.search(r'(in the last \d+ months)|(lifetime)', text)
    if time:
        return time.group(1) if time.group(1) else time.group(2)
    else:
        return None
df_alter['rating_period'] = df_alter['sellerratings'].apply(extract_time)
```

```python
def extract_num_ratings(text):
    num_ratings = re.search(r'\((\d+) ratings\)|\((\d+) total ratings\)', text)
    if num_ratings:
      return num_ratings.group(1) if num_ratings.group(1) else num_ratings.group(2)
    else:
      return None
df_alter['number_of_ratings'] = df_alter['sellerratings'].apply(extract_num_ratings)
```

### Scoring System

```python
def calculate_score(row):
    # Define weights
    weight_positive_percentage = 0.4
    weight_num_brands = 0.1
    weight_max_negative_30_days = 0.2
    weight_max_negative_90_days = 0.2
    weight_max_negative_12_months = 0.4
    weight_hero_products = 0.1  # (combined weight for both hero products)

    # Normalize values
    normalized_positive_percentage = row['positive_percentage_ratings'] / 100
    normalized_num_brands = row['Count of seller brands'] / 16  # Assuming max is 16
    normalized_max_negative_30_days = row['Max % of negative seller ratings - last 30 days'] / 100
    normalized_max_negative_90_days = row['Max % of negative seller ratings - last 90 days'] / 100
    normalized_max_negative_12_months = row['Max % of negative seller ratings - last 12 months'] / 100
    normalized_hero_products = (row['Hero Product 1 #ratings'] + row['Hero Product 2 #ratings']) / 100000  # Normalize for scale

    # Calculate score
    score = (
        weight_positive_percentage * normalized_positive_percentage +
        weight_num_brands * normalized_num_brands +
        weight_max_negative_30_days * (1 - normalized_max_negative_30_days) +  # Invert the negative rating
        weight_max_negative_90_days * (1 - normalized_max_negative_90_days) +
        weight_max_negative_12_months * (1 - normalized_max_negative_12_months) +
        weight_hero_products * normalized_hero_products
    )

    return score
```

```python
df_alter['seller_score'] = df_alter.apply(calculate_score, axis=1)
```

```python
df_ranked = df_alter.sort_values(by='seller_score', ascending=False)
```
